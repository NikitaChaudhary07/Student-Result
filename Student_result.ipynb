```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=df.drop("Unnamed: 0", axis=1)
df
```

# GENDER DISTURIBUTION

```python
colors =['green', 'yellow']
ab=sns.countplot(x="Gender", data=df,palette=colors,hue='Gender')
ab.bar_label(ab.containers[0])
ab.bar_label(ab.containers[1])
plt.title("Gender Disturibution")
plt.show()
```

#From Gender Disturibution chart we can analyse that Females are more than in Males in Result

# Parent education and Qualification Impact

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("Impact of Parent's Education on Student's Score")
plt.show()
```

# This map shows that the  parents education has good impact on their Childern Result .
# As Children whose parents has completed Bachelor's or Master's degree has scored well than others.
# Top is of Master's degree parents than Bachelor's one.

# Parent Maritual Status Impact

```python
gb = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("Impact of Parent's MaritalStatus on Student's Score")
plt.show()
```

# This map shows that the  Parents MaritalStatus have no impact or negligible impact on their Children Result .
# As there is not much difference in average scores of children scores .

# Weekly Hour Study Impact on Scores

```python
gb = df.groupby("WklyStudyHours").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
gb.plot(kind='line')
plt.title("Average Scores by Weekly Study Hours")
plt.xlabel("Weekly Study Hours")
plt.ylabel("Average Score")
plt.grid(False)
plt.legend(title="Subjects")
plt.tight_layout()
plt.show()
```

# Shows that who study more than 10 hours has scored better than those who study for 5-10 hours,
# Also who study 5-10 hour have scored well than who study less than 5 hours

# Test Preparation Impact on Score

```python
gb = df.groupby("TestPrep").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=[8,8])
ax = gb.plot(kind='bar', color=['yellow', 'lightblue','red'])
for container in ax.containers:
    ax.bar_label(container)
plt.title("Test Preparation Disturibution")
plt.show()
```

# This graph shows that who have completed their test preparation have scored well than those who does not completed it.

# Gender wise Test Preparation Disturibution

```python
ax = sns.countplot(x='Gender', hue='TestPrep', data=df, palette=['yellow', 'red'])


for container in ax.containers:
    ax.bar_label(container)

plt.title("Test Preparation by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Test Preparation")
plt.show()
```

# Number of students who did not completed the test preparation is almost twice the no. of student who completed their preparation.
# More female Students completed their preparation than males.
